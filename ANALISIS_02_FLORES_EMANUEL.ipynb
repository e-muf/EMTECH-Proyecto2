## An√°lisis por Rutas (Importaci√≥n y Exportaci√≥n)

```python
count_directions = df.groupby(['direction', 'origin', 'destination']).agg({'origin':['count'],'total_value':['sum']})
count_directions.columns = ['routes_count', 'sum_total_value']

imports_count = count_directions.query('direction == "Imports"')
exports_count = count_directions.query('direction == "Exports"')

imports_count = imports_count.sort_values(by=['routes_count'], ascending=False)
exports_count = exports_count.sort_values(by=['routes_count'], ascending=False)
```

### Cantidad de Exportaciones por cada pa√≠s

```python
country_direction = pd.crosstab(df['origin'], df['direction'])
country_order_by_exports = country_direction.sort_values(by="Exports", ascending=False)
country_order_by_exports.Exports.plot.bar(figsize=(15, 5))
plt.title(label="Exportaciones por cada pa√≠s")
plt.show()
```

### Cantidad de Importaciones por cada pa√≠s

```python
country_order_by_imports = country_direction.sort_values(by="Imports", ascending=False)
country_order_by_imports.Imports.plot.bar(figsize=(15, 5), color=['#ff7f0e'])
plt.title(label="Importaciones por cada pa√≠s")
plt.show()
```

## An√°lsis por Valor de Importaciones y Exportaciones

```python
country_values = df.groupby(['direction', 'origin']).agg({'total_value':['sum']})
country_values.columns = ['total_value_country']

country_exports_value = country_values.query('direction == "Exports"').sort_values(by=['total_value_country'], 
                                                                                   ascending=False)
country_imports_value = country_values.query('direction == "Imports"').sort_values(by=['total_value_country'], 
                                                                                   ascending=False)
```

```python
total_exports_value = sum(country_exports_value.total_value_country)
country_exports_value['mean_values'] = country_exports_value.total_value_country / total_exports_value
country_exports_value
```

```python
total_imports_value = sum(country_imports_value.total_value_country)
country_imports_value['mean_values'] = country_imports_value.total_value_country / total_imports_value
country_imports_value
```

```python
country_exports_value.mean_values.plot.bar(figsize=(15,5))
plt.title('Promedio del valor de exportaciones por pa√≠s')
plt.show()
```

```python
country_imports_value.mean_values.plot.bar(figsize=(15,5), color=['#ff7f0e'])
plt.title('Promedio del valor de importaciones por pa√≠s')
plt.show()
```

```python
total_values = total_exports_value + total_imports_value
country_total_values = df.groupby(['origin']).agg({'total_value':['sum']})
country_total_values['mean_values'] = country_total_values.total_value / total_values
country_total_values = country_total_values.sort_values(by=['mean_values'])
country_total_values.mean_values.plot.bar(figsize=(15,5), color=['#ab47bc'])
plt.title('Promedio del valor de exportaci√≥n e importaci√≥n por pa√≠s')
plt.show()
```

```python
country_total_values.columns = ['total_value', 'mean_value']
eighty_percent = country_total_values.total_value.quantile(0.8)
```

## An√°lisis por Transporte
